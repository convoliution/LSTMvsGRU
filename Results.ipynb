```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# results
#
#       | sherlock | shakespeare | coldplay |
# ------+----------+-------------+----------|
#  LSTM |   [0,0]  |    [0,1]    |   [0,2]  |  
# ------+----------+-------------+----------|
#  GRU  |   [1,0]  |    [1,1]    |   [1,2]  |  
# ------+----------+-------------+----------|

depth = np.load("results/loss-histories/LSTM-Sherlock.npy").size
results = np.zeros((2, 3, depth))

results[0,0] = np.load("results/loss-histories/LSTM-Sherlock.npy")
results[0,1] = np.load("results/loss-histories/LSTM-Shakespeare.npy")
results[0,2] = np.load("results/loss-histories/LSTM-Coldplay.npy")
results[1,0] = np.load("results/loss-histories/GRU-Sherlock.npy")
results[1,1] = np.load("results/loss-histories/GRU-Shakespeare.npy")
results[1,2] = np.load("results/loss-histories/GRU-Coldplay.npy")
```

```python
plt.plot(results[0,0], 'C0-', lw=.5, label='LSTM Sherlock')
plt.plot(results[0,1], 'C1-', lw=.5, label='LSTM Shakespeare')
plt.plot(results[0,2], 'C2-', lw=.5, label='LSTM Coldplay')
plt.plot(results[1,0], 'C0:', label='GRU Sherlock')
plt.plot(results[1,1], 'C1:', label='GRU Shakespeare')
plt.plot(results[1,2], 'C2:', label='GRU Coldplay')
plt.title("Loss Histories")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.legend(loc='best')
plt.savefig("loss.png")
```
